# **Text cleaning**

```python
!pip install nltk
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

def remove_url(text):
    text = str(text)  # Convert 'text' to string if it's not already
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

def remove_punc(text):
    text = str(text)  # Convert 'text' to string if it's not already
    return text.translate(str.maketrans('', '', exclude))

def remove_stopwords(text):
    text = str(text)  # Convert 'text' to string if it's not already
    stop_words = set(stopwords.words('english'))  # Get set of English stop words
    words = text.split()  # Split text into words
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Filter out stop words
    return ' '.join(filtered_words)  # Join filtered words back into a string

df['tweet'] = df['tweet'].str.lower()
df['tweet'] = df['tweet'].apply(remove_url)
df['tweet'] = df['tweet'].apply(remove_punc)
df['tweet'] = df['tweet'].apply(remove_stopwords)  # Apply stop word removal
```

# spliting dataset

```python
import tensorflow as tf
import tensorflow_hub as hub
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.callbacks import EarlyStopping
# Clear the session to avoid resource conflicts
tf.keras.backend.clear_session()
# Assuming 'df' is the DataFrame containing your data
X = df['tweet']  # Text column
Y = df['sentiment']  # Label column

# Encode the labels (Y)
encoder = LabelEncoder()
Y = encoder.fit_transform(Y)  # Converts labels to 0 and 1

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Define BERT model for fine-tuning
preprocessor = hub.KerasLayer("https://kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/en-uncased-preprocess/versions/3")
encoder = hub.KerasLayer("https://www.kaggle.com/models/tensorflow/bert/frameworks/TensorFlow2/variations/en-uncased-l-12-h-768-a-12/versions/4", trainable=True)

# Input layer
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
encoder_inputs = preprocessor(text_input)
outputs = encoder(encoder_inputs)

# Pooled output
pooled_output = outputs["pooled_output"]

# Dropout and classification layers
drop_out = tf.keras.layers.Dropout(0.3, name='dropout')(pooled_output)  # Regularization
output = tf.keras.layers.Dense(1, activation='sigmoid', name='classifier')(drop_out)  # Binary classification (1 unit with sigmoid)

# Build the model
model = tf.keras.Model(inputs=[text_input], outputs=[output])
```

```python
# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5),
    loss='binary_crossentropy',  # Use binary_crossentropy for binary classification
    metrics=['accuracy']
)
```
